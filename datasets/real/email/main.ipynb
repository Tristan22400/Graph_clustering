```python
import networkx as nx

g = nx.read_edgelist("email.edgelist")
nx.write_gml(g, "email.gml")
g = nx.read_gml("email.gml")
```

```python
with open("email.coms", "r") as f:
    coms = f.readlines()
```

```python
for node_com in coms:
    node = node_com.split()[0]
    com = node_com.split()[1]
    g.nodes[node]["value"] = com
```

```python
nx.write_gml(g, "email.gml")
```
